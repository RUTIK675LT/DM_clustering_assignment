```python
!pip install transformers torch

import torch
from transformers import BertModel, BertTokenizer
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np

# Function to create embeddings using BERT
def get_bert_embeddings(text, tokenizer, model):
    inputs = tokenizer(text, return_tensors="pt", max_length=512, truncation=True)
    outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).detach().numpy()

# Load the BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Load your dataset
url = '/content/document_clustering_sample_data.csv'  # Replace with the path to the CSV file
df = pd.read_csv(url)

# Generate embeddings
embeddings = np.vstack(df['Document'].apply(lambda x: get_bert_embeddings(x, tokenizer, model)))

# Perform k-means clustering
n_clusters = 3  # Adjust as needed
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
labels = kmeans.fit_predict(embeddings)

# Add cluster labels to the DataFrame
df['Cluster'] = labels

# Analyze the clusters
for i in range(n_clusters):
    print(f"Cluster {i} documents:")
    print(df[df['Cluster'] == i]['Document'], "\n")
```
